## Task 1. Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
```

## Task 2. Loading Dataset for Analysis

```python
df = pd.read_csv('weatherhistory.csv')
```

```python
df.head()
```

## Task 3. Creating DataFrame of Useful Columns 

```python
cols = ['Formatted Date', 'Apparent Temperature (C)', 'Humidity']
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'] , utc = True)
```

```python
new_df = df[cols]
new_df.head()
```

## Task 4. Resampling Our DataFrame 

```python
new_df = new_df.set_index('Formatted Date')
new_df = new_df.resample('MS').mean()
```

```python
new_df.head()
```

## Task 5. Resampled DataFrame With Apparent Temp.

```python
plt.figure(figsize = (15,7))

style.use('ggplot')

plt.plot(new_df.index , new_df['Apparent Temperature (C)'], 'ro-')
plt.plot(new_df.index , np.log(new_df.Humidity) , 'go-')

plt.title("Apparent Temperature & Humidity With Time ",fontsize = 18)                       
plt.xlabel('Timeline', fontsize = 18)                              
plt.ylabel('Apparent Temperature & Humidity', fontsize = 17)                                     
plt.grid(color = 'k', linestyle = '--', linewidth = 0.9)              
plt.legend(['Temperature','Humidity'] , loc = 1)

plt.show()
```

## Task 6. Analyze Conditions in Month-wise
