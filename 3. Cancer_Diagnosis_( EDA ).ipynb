## Task 1: Loading Libraries and Data

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns  # data visualization library  
import matplotlib.pyplot as plt
import time
```

```python
df1 = pd.read_csv('data.csv')
df1.head()
```

## Task 2: Separate Target from Features

```python
df1.columns
```

```python
drop_cols = ['Unnamed: 32','id','diagnosis']
df2 = df1.drop(drop_cols, axis = 1)
df2.head()
```

## Task 3: Plot Diagnosis Distributions

```python
plt.figure(figsize = (9,5))
a = df1.diagnosis
sns.countplot(a,palette = 'hot')
B,M = a.value_counts()
print('No. of Benign Tumors',B)
print('No. of Malignant Tumors',M)
plt.title('Counts of Tumors',fontsize = 18)
plt.show()
```

```python
df2.describe()
```

## Task 4: Visualizing Standardized Data with Seaborn

```python
plt.figure(figsize = (9,5))
df2_std = (df2 - df2.mean()) / df2.std()
df2 = pd.concat([a,df2_std.iloc[:,0:10]], axis = 1)
df2 = pd.melt(df2, id_vars = 'diagnosis', var_name = 'features', value_name = 'values')

df2
```

## Task 5: Violin Plots and Box Plots

```python
plt.figure(figsize = (10,6))
df2 = pd.concat([a,df2_std.iloc[:,0:10]], axis = 1)
df2 = pd.melt(df2, id_vars = 'diagnosis', var_name = 'features', value_name = 'values')

sns.violinplot(x = 'features',y = 'values',hue = 'diagnosis', data = df2, split = True,inner = 'quartile')
plt.xticks(rotation = 45)
plt.title('Initial 10 Features Of Diagnosis',fontsize = 18)
plt.show()
```

```python
plt.figure(figsize = (10,6))
df2 = pd.concat([a,df2_std.iloc[:,10:20]], axis = 1)
df2 = pd.melt(df2, id_vars = 'diagnosis', var_name = 'features', value_name = 'values')

sns.violinplot(x = 'features',y = 'values',hue = 'diagnosis', data = df2, split = True,inner = 'quartile')
plt.xticks(rotation = 40)
plt.title('Next 10 Features Of Diagnosis',fontsize = 18)
plt.show()
```

```python
plt.figure(figsize = (12,7))
sns.boxplot(x = 'features', y = 'values',hue = 'diagnosis',data = df2,palette = 'Dark2')
plt.xticks(rotation = 45)
plt.title('Values v/s Features',fontsize = 22)
plt.show()
```

**ABOUT DATA SUMMARY**

Ten real-valued features are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter) 
2. texture (standard deviation of gray-scale values) 
3. perimeter 
4. area 
5. smoothness (local variation in radius lengths) 
6. compactness (perimeter^2 / area - 1.0) 
7. concavity (severity of concave portions of the contour) 
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension ("coastline approximation")

All feature values are recoded with four significant digits.

Missing attribute values: none

Class distribution: 357 benign, 212 malignant
