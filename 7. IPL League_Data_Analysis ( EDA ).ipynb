# <center> IPL League Data Analysis : ( EDA ) 

## Task 1. Import Libraries  

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
```

## Task 2. Import  1st Dataset  

```python
data = pd.read_csv('deliveries.csv')
```

```python
data.head(3)
```

## Task 3. Plotting Balls v/s Runs 

```python
plt.figure(figsize = (10,6))
style.use('ggplot')                                                 
plt.plot('ball','total_runs',data = data.head(18), color = 'g', linestyle = '--')
plt.title("Total Balls v/s Runs ",fontsize = 15)                    
plt.xlabel('Balls', fontsize = 15)                     
plt.ylabel('Total Runs', fontsize = 15)                                 
plt.legend(['Scores'], loc = 1)                                       
plt.grid(color = 'k', linestyle = '--', linewidth = 0.7)               
plt.show() 
```

## Task 4. Analyze Max Overs Occured In Season

```python
over = data[data['match_id']== 1]['over']
runs = data[(data['match_id']==1) & (data['over']==1)]['total_runs'].sum()
```

```python
(data['over'].value_counts() / 6).plot(kind = 'bar',figsize = (16,7)
                                      ,color = ['c','b','w','k','m','r','g','y']
        , edgecolor = 'k', linewidth = 2,alpha = 0.75)
plt.title("Over-wise Balls Delivered ",fontsize = 22)                    
plt.xlabel('Overs ', fontsize = 15)                     
plt.ylabel('Balls Delivered ', fontsize = 15)                                 
plt.legend(['Total Balls'], loc = 1)                                       
plt.grid(color = 'k', linestyle = '--', linewidth = 0.4) 
plt.show()
```

## Task 5. Import 2nd Dataset  

```python
match = pd.read_csv('matches.csv')
```

```python
match.head(3)
```

## Task 6. Analyze Most Man Of The Match Award Winners

```python
mom = match['player_of_match'].value_counts().head(10)
```

```python
mom.plot(kind = 'pie',figsize = (14,8),autopct = '%0.1f%%', textprops = {'fontsize':13}
       , wedgeprops = {'linewidth':2,'width':1,'edgecolor':'k'},shadow = True)
plt.legend(loc = 3)
plt.show()
```

## Task 7. Analyze Most No. of Wins Team-wise

```python
win = match['winner'].value_counts()
```

```python
win.plot(kind = 'bar', figsize = (12,7), color = ['c','b','w','k','m','r','g','y']
        , edgecolor = 'k', linewidth = 2,alpha = 0.8)
plt.title("Most Times Winner ",fontsize = 22)                    
plt.xlabel('Teams ', fontsize = 17)                     
plt.ylabel('No. of Times ', fontsize = 17)                                 
plt.legend(loc = 1,fontsize = 18)                                       
plt.grid(color = 'k', linestyle = '-', linewidth = 0.4)
plt.show()
```
